```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D, Dense, Dropout, Flatten
from keras.optimizers import Adam

datadir = '/content/drive/MyDrive/3RD_SEM_PROJ'
img_dir = os.path.join(datadir, 'IMG')
csv_path = os.path.join(datadir, 'driving_log.csv')
print('img_dir =', img_dir)
print('csv_path =', csv_path)
```

```python
columns = ['center','left','right','steering','throttle','reverse','speed']
dataset = pd.read_csv(csv_path, names=columns)
import ntpath
def removePath(path):
    _, tail = ntpath.split(path)
    return tail
dataset['center'] = dataset['center'].apply(removePath)
dataset['left']   = dataset['left'].apply(removePath)
dataset['right']  = dataset['right'].apply(removePath)
```

```python
num_bins = 25
hist, bins = np.histogram(dataset['steering'], num_bins)
threshold = 500
remove_list = []
for i in range(num_bins):
    idxs = []
    for j in range(len(dataset['steering'])):
        if dataset['steering'][j] >= bins[i] and dataset['steering'][j] <= bins[i+1]:
            idxs.append(j)
    idxs = shuffle(idxs)
    idxs = idxs[threshold:]
    remove_list.extend(idxs)
dataset.drop(dataset.index[remove_list], inplace=True)
```

```python
def loadImageSteering(datadir, dataset):
    imagePath = []
    steeringPath = []
    for i in range(len(dataset)):
        center = dataset.iloc[i][0]
        steering = float(dataset.iloc[i][3])
        imagePath.append(os.path.join(datadir, 'IMG', center))
        steeringPath.append(steering)
    return np.asarray(imagePath), np.asarray(steeringPath)
imagePath, steeringPath = loadImageSteering(datadir, dataset)
```

```python
x_train, x_test, y_train, y_test = train_test_split(imagePath, steeringPath, test_size=0.2, random_state=6)
```

```python
def imagePreprocessing(img_path):
    img = mpimg.imread(img_path)
    img = img[60:135, :, :]
    img = cv2.cvtColor((img * 255).astype(np.uint8), cv2.COLOR_RGB2YUV).astype(np.float32) / 255.0
    img = cv2.GaussianBlur(img, (3,3), 0)
    img = cv2.resize(img, (200, 66))
    return img
```

```python
idx = 0
orig = mpimg.imread(imagePath[idx])
proc = imagePreprocessing(imagePath[idx])
fig, axs = plt.subplots(1,2,figsize=(12,6))
axs[0].imshow(orig)
axs[0].axis('off')
proc_rgb = cv2.cvtColor((proc*255).astype(np.uint8), cv2.COLOR_YUV2RGB)
axs[1].imshow(proc_rgb)
axs[1].axis('off')
plt.show()
```

```python
x_train_proc = np.array(list(map(imagePreprocessing, x_train)))
x_test_proc  = np.array(list(map(imagePreprocessing, x_test)))
```

```python
def nvidiaModel():
    model = Sequential()
    model.add(Conv2D(24, (5,5), strides=(2,2), input_shape=(66,200,3), activation="elu"))
    model.add(Conv2D(36, (5,5), strides=(2,2), activation="elu"))
    model.add(Conv2D(48, (5,5), strides=(2,2), activation="elu"))
    model.add(Conv2D(64, (3,3), activation="elu"))
    model.add(Conv2D(64, (3,3), activation="elu"))
    model.add(Dropout(0.5))
    model.add(Flatten())
    model.add(Dense(100, activation="elu"))
    model.add(Dropout(0.5))
    model.add(Dense(50, activation="elu"))
    model.add(Dropout(0.5))
    model.add(Dense(10, activation="elu"))
    model.add(Dropout(0.5))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=1e-3), loss="mse")
    return model
model = nvidiaModel()
model.summary()
```

```python
h = model.fit(x_train_proc, y_train, validation_data=(x_test_proc, y_test), epochs=30, batch_size=100, shuffle=True, verbose=1)
plt.plot(h.history['loss'], label='loss')
plt.plot(h.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.models import load_model
model = load_model('/content/drive/MyDrive/3RD_SEM_PROJ/model.keras')
def preprocess_for_inference(img_path):
    img = mpimg.imread(img_path)
    img = img[60:135, :, :]
    img = cv2.cvtColor((img * 255).astype(np.uint8), cv2.COLOR_RGB2YUV).astype(np.float32) / 255.0
    img = cv2.GaussianBlur(img, (3,3), 0)
    img = cv2.resize(img, (200, 66))
    img = np.expand_dims(img, axis=0)
    return img
```

```python
valid_ext = (".jpg", ".jpeg", ".png", ".bmp")
files = [f for f in os.listdir(img_dir) if f.lower().endswith(valid_ext)]
random_img = os.path.join(img_dir, random.choice(files))
img_input = preprocess_for_inference(random_img)
angle = model.predict(img_input)[0][0]
orig = mpimg.imread(random_img)
fig = plt.figure(figsize=(14,5))
ax_img = fig.add_subplot(1,2,1)
ax_img.imshow(orig)
ax_img.axis('off')
fig.text(0.99, 0.5, f"Predicted Steering Angle\n{angle:.4f}", ha='right', va='center', fontsize=28)
plt.show()
```
